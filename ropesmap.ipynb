```python
import pandas as pd
import numpy as np
import requests
import folium
import json
import os
import pathlib as pl
from IPython.display import display
from folium import plugins
from folium.plugins import MarkerCluster
import branca.colormap as cm
```

```python
df = pd.read_csv("/content/shibari_places.csv")
```

```python
df.head()
```

```python
m = folium.Map(location = [41.29246, 12.5736108], zoom_start=6, tiles = None)

t = folium.TileLayer(tiles="cartodbdark_matter", control=False).add_to(m)
```

```python
mcg = folium.plugins.MarkerCluster(control=False)
m.add_child(mcg)
fg = folium.FeatureGroup(name = "My Map")
```

```python
#novità!
from urllib.parse import urlparse
for index, row in df.iterrows():
        contatti = f"{row['Contatti']}"
        riferimento1 = f"{row['IG_riferimento_1']}"
        riferimento2 = f"{row['IG_riferimento_2']}"
        riferimento3 = f"{row['IG_riferimento_3']}"
        url = f"{row['Icona']}"
        icona = folium.features.CustomIcon(url, icon_size=(30,30))
        html = f'''

        <center><b><h2>{row['Nome']}</h2></b>
        <p><b>{row['Citta']}</b></center>
        <p><b>Descrizione</b><br>{row['Descrizione']}
        <p><b>Descrizione</b><br>{row['Stile']}
        <p><b>Contatti</b>: <a href={contatti} target="_blank">Instagram</a>
        <p><b>Persone di riferimento</b>: <a href={riferimento1} target="_blank">{row['Riferimento_1']}</a>
        <br><a href={riferimento2} target="_blank">{row['Riferimento_2']}</a>
        <br><a href={riferimento3} target="_blank">{row['Riferimento_3']}</a>
        '''
        folium.Marker([row['Latitudine'], row['Longitudine']],
        popup = folium.Popup(html,
                 min_width=200,
                 max_width=200),
        tooltip=row['Nome'],
        icon = icona
                    ).add_to(mcg)
```

```python
m.add_child(fg)
```

```python
m.save(open("mappa_ropes.html", "wb"))
```
